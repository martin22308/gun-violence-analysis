# **Feature Selection**

```python
data=df[['date','state','city_or_county','n_killed','n_injured']]
data=data.dropna()#هيشيل كل الصفوف الnan
data
```

# **Time Feature Engineering**

```python
data['date'] = pd.to_datetime(data['date'])
data = data[data['date'].notna()]
```

```python
data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month
data['day'] = data['date'].dt.day
data['day_of_week'] = data['date'].dt.dayofweek
```

```python
data['severity'] = data['n_killed'] + data['n_injured']
```

```python
data['day_of_week'] = data['day_of_week'].map({0: 'Monday', 1: 'Tuesday',
                                               2: 'Wednesday', 3: 'Thursday',
                                               4: 'Friday', 5: 'Saturday', 6: 'Sunday'})
data.head()
```

# **Day of Week Analysis**
Identify which days of the week experience higher gun violence incidents and severity

```python
incidents_by_day = data.groupby('day_of_week').size().sort_values(ascending=False)
incidents_by_day
```

```python
plt.figure(figsize=(8,4))
incidents_by_day.plot(kind='bar')
plt.title("Number of Incidents by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Number of Incidents")
plt.show()
```

# **Severity Analysis by Day of Week**

```python
severity_by_day = data.groupby('day_of_week')['severity'] \
    .agg(mean='mean', median='median', max='max') \
    .sort_values(by='max', ascending=False)

severity_by_day
```
